```python
def impute_age(cols):
    Age = cols.iloc[0]
    Pclass= cols.iloc[1]

    if pd.isnull(Age):
        
        if Pclass == 1:
                return 37

        elif Pclass ==2:
                return 29

        else:
            return 24

    else:
          return Age
```

```python
train['Age']=train[['Age','Pclass']].apply(impute_age,axis=1)
```

```python
sex=pd.get_dummies(train['Sex'],drop_first=True)  #We'll need to convert categorical features to dummy variables using pandas! Otherwise our machine learning algorithm won't be able to directly take in those features as inputs.
embarked = pd.get_dummies(train['Embarked'],drop_first=True)
```

```python
train=pd.concat([train,sex,embarked],axis=1)
```

```python
train.head()
```

```python
train.drop(['Name','Sex','Ticket','Embarked'],axis=1,inplace=True)
```

```python
train.head()
```

```python
from sklearn.model_selection import train_test_split    # there is a separate test dataset in the folder if want to use this full dataset for training instead of splitting it
```

```python
X_train,X_test,y_train,y_test= train_test_split(train.drop('Survived',axis=1),train['Survived'],test_size=0.3,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train,y_train)
```

```python
predix= logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predix))
```
